```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split,cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
cancer = load_breast_cancer()
```

```python
X = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
y = cancer['target']
```

```python
scaler_x = StandardScaler().fit_transform(X)
```

```python
from sklearn.svm import SVC
```

```python
Xr, Xt, yr, yt = train_test_split(X, y, test_size=0.3, random_state = 0)
```

```python
model = SVC().fit(Xr, yr)
```

```python
pred = model.predict(Xt)
```

```python
import numpy as np
np.mean(cross_val_score(model, Xt, yt, cv = 3, scoring = 'accuracy'))
```

```python
np.mean(cross_val_score(model, Xt, yt, cv = 3, scoring = 'f1_weighted'))
```

```python
from sklearn.metrics import classification_report

print(classification_report(yt, pred))
```

```python
params = {'C':[1,0.1,0.01,0.001],'gamma':[1,10,100,1000], 'kern'}
```

```python
grid = GridSearchCV(SVC(), params, scoring = 'accuracy', cv = 3, verbose = 5)
```

```python
model2 = grid.fit(Xr, yr)
```

```python
model2.best_estimator_
```
