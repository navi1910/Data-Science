```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('loan_data.csv')
```

```python
df
```

```python
df.info()
```

```python
df['purpose'].value_counts()
```

```python
df = pd.concat([df,pd.get_dummies(df['purpose'])], axis = 1).drop('purpose', axis = 1)
```

```python
df
```

```python
plt.figure(figsize = (15,15))
sns.heatmap(df.corr(), annot = True)
```

```python
df.corr()['not.fully.paid']
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100)
```

```python
y = df['not.fully.paid']
```

```python
X = df.drop('not.fully.paid', axis = 1)
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

Xtr, Xts, ytr, yts = train_test_split(X, y, random_state=0, test_size = 0.3)
```

```python
model.fit(Xtr, ytr)
```

```python
model.score(Xts, yts)
```

```python
df['not.fully.paid'].value_counts()
```

```python
pred = model.predict(Xts)
```

```python
print(confusion_matrix(pred, yts))
print(classification_report(pred, yts))
```

```python
btes = models.best_estimator_
```

```python
models.best_score_
```
