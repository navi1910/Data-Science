```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.to_csv('Tips.csv')
```

```python
sns.set_style('white')
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(tips['total_bill'], color = 'red')
plt.title('Histogram')
plt.show()
```

```python
sns.distplot(tips['total_bill'], color = 'red', kde = False)
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(tips['total_bill'], color = 'red', kde = False)
plt.title('Histogram')
plt.xlabel('Total Bill')
plt.ylabel('The Density')
plt.show()
```
