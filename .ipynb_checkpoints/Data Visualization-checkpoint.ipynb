```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.to_csv('Tips.csv')
```

```python
tips.info()
```

```python
sns.set_style('white')
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(tips['total_bill'], color = 'red')
plt.title('Histogram')
plt.show()
```

```python
sns.distplot(tips['total_bill'], color = 'red', kde = False)
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(tips['total_bill'], color = 'red', kde = False)
plt.title('Histogram')
plt.xlabel('Total Bill')
plt.ylabel('The Density')
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
tips[['total_bill', 'tip']].corr()
```

```python
tips['day'].unique()
sun = tips[tips['day'] == 'Sun']
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = sun, hue = 'sex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind='kde')
```

```python
plt.scatter(x = 'total_bill', y = 'tip', c= 'r', data = tips, marker = 'x', alpha = 0.50)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex', palette='Paired')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x= 'day', y = 'total_bill', data = tips)
```
